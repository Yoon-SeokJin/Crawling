```python
import pandas as pd
import numpy as np
def get_hdf_keys(dir):
    with pd.HDFStore(dir) as store:
        return [e[1:] for e in store.keys()]
dir = [
    'danawa_crawling_data_cmoh.h5',
    'danawa_crawling_data_seokjin.h5',
    'danawa_crawling_data_jinu.h5',
]
key_list = [get_hdf_keys(e) for e in dir]
key_combine = list(set([e for k in key_list for e in k]))
key_id = [e for e in key_combine if not e.endswith('review') and not e.endswith('detail')]
key_detail = [e for e in key_combine if e.endswith('detail')]
key_review = [e for e in key_combine if e.endswith('review')]
key_id.sort()
key_detail.sort()
key_review.sort()
key_id, key_detail, key_review
combine_dir = 'danawa_crawling_data_combined.h5'
for key in key_id:
    df1 = pd.read_hdf(dir[0], key)
    df2 = pd.read_hdf(dir[1], key)
    df3 = pd.read_hdf(dir[2], key)
    df1['id_validator'] = df1['id_validator'].astype(np.int32)
    df2['id_validator'] = df2['id_validator'].astype(np.int32)
    df3['id_validator'] = df3['id_validator'].astype(np.int32)
    dfc = df1.combine_first(df2.combine_first(df3))
    dfc.to_hdf(combine_dir, key)
```

```python
rename_dict = {'이름':'name', '이미지링크':'image', '최저가로고':'shop_logo', '최저가브랜드명':'shop_name', '최저가링크':'shop_link'}

for k in key_id[0:1]:
    df0 = pd.read_hdf(combine_dir, k)
    df0 = df0[df0['id_validator'] == True]
    df0.drop(['id_validator'], axis=1, inplace=True)
    df_comb = None
    for i in range(3):
        if k + '_detail' in key_list[i]:
            df = pd.read_hdf(dir[i], k + '_detail')
            df.rename(columns=rename_dict, inplace=True)
            print(df.columns)
            if df_comb is not None:
                df_comb = df_comb.merge(df, left_index=True, right_index=True, suffixes=['', '_x'])
                df_comb.drop(columns=[col for col in df_comb.columns if col.endswith('_x')], inplace=True)
            else:
                df_comb = df
    if df_comb is None:
        continue
    df0 = df0.merge(df_comb, how='right', left_index=True, right_index=True)
    df0.to_hdf(combine_dir, k + '_detail')
```

```python
for i in range(3):
    for key in key_list[i]:
        if key.endswith('review') and key not in [y for x in key_list[0 : i] for y in x]:
            df = pd.read_hdf(dir[i], key)
            df.to_hdf(combine_dir, key)
```
